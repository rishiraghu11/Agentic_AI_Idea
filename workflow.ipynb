```python
import pandas as pd
```

```python
df = pd.read_excel("vehicle_dataset.xlsx")
df
#Code to import dataset
```

```python
# Rule-based failure prediction logic for prototype demonstration
# Uses predefined thresholds on synthetic vehicle telematics data
def predict_failure(row):
    if row['engine_temp'] > 100:
        return "Engine Overheating", "High"
    elif row['brake_wear_pct'] > 80:
        return "Brake Wear", "High"
    elif row['vibration'] > 0.75:
        return "Bearing / Vibration Issue", "Medium"
    else:
        return "Normal", "Low"
```

```python
# Apply failure prediction logic to each vehicle record
# and generate predicted issue and risk level autonomously
df[['predicted_issue', 'risk_level']] = df.apply(
    lambda row: pd.Series(predict_failure(row)), axis=1
)

df
```

```python
# Service recommendation logic for the prototype
# Converts predicted risk levels into actionable maintenance decisions
# to demonstrate autonomous decision-making in the Agentic AI workflow
def recommend_service(risk):
    if risk == "High":
        return "Immediate service recommended (within 48 hours)"
    elif risk == "Medium":
        return "Schedule service within next 7 days"
    else:
        return "No immediate service required"
```

```python
# Generate autonomous service recommendations for each vehicle
# based on the predicted risk level
df['service_action'] = df['risk_level'].apply(recommend_service)

df
```

```python
import sqlite3
```

```python
# Create a lightweight SQLite database to store vehicle health data
# for SQL-based prioritization and analysis in the prototype
conn = sqlite3.connect("vehicle_health.db")
```

```python
df.to_sql("vehicle_health", conn, if_exists="replace", index=False)
```

```python
# SQL query to identify and prioritize high-risk vehicles
# requiring immediate service intervention
query = """
SELECT vehicle_id,
       predicted_issue,
       risk_level,
       service_action
FROM vehicle_health
WHERE risk_level = 'High'
"""
```

```python
# Execute SQL query to retrieve high-risk vehicles
# and display the prioritized service list
high_risk_df = pd.read_sql(query, conn)
high_risk_df
```
